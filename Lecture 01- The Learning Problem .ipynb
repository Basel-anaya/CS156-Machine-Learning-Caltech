## Components of Learning



### Formalization:

1. **Input:** *X*
- We call the input data as “X” input data is same as feature vector. In this case it can be the customer vector( just to make it clear , vector is nothing but the collection of attributes describing the entity such as movie / customer ).
2. **Output:** *Y*
- The output is called “Y”. This is distinct output i.e ± 1.
3. **Target function:** *f* **:** *X* &rarr; *Y*
- The target function f: X ->Y indicates the ideal credit formula.i.e whether the person should be given credit or not. (Read it as” a function f such that X implies Y”).
4. **Data:** (*x*<sub>*1*</sub>, *y*<sub>*1*</sub>),(*x*<sub>*2*</sub>, *y*<sub>*2*</sub>),**....**, (*x*<sub>*N*</sub>, *y*<sub>*N*</sub>)
- The Data consists of all customers such as( x1,y1) ,(x2,y2) ..(xn;yn)

**&darr;** **&darr;** **&darr;**

5. Hypothesis: *g* **:** *X* **&rarr;** *Y*
- We also define a Hypothesis Set g: X->Y. The hypothesis set if defined is a list of all possible functions that helps us achieve the target function .

Notes:

- The **Target Function** is unknown to us, if it was known then there's no need for learning.

- **Hypothesis** is the formal name of the formula we are going to use to approximate the **Target function** == *g* in the **Hypothesis** approximates *f*

**So: There is no control on the target function, no control on the input or training examples. The only things we can control are the learning algorithm (LA) and the hypothesis set (HS) . Remember LA+HS are our learning tools . All of machine learning we use the hypothesis set and a learning algorithm along with it to decide on the target function e.g Hypothesis set can be Neural network and the learning algorithm can be a back propagation neural network or a feed forward neural network.**

### Our first ML algorithm ( the most rudimentary ever built) Perceptron classifier for credit approval.

#### The approve credit function can be defined mathematically as follows:

- Approve Credit if Σi=1 d wixi > Threshold

- Deny Credit if Σi=1 d wixi < Threshold { where w is the weight for each attribute. i.e salary is high so weight is more positive. Customer has lot of outstanding loans to pay so weight is negative etc.}

So , a little modification to the equation

h(x) = sign((Σi=1 d wixi ) – Threshold) …..1

= sign((Σi=1 d wixi )  + w0)  …………..2  { where w0 is = -Threshold}

= sign (Σi=0 d wixi ) ……………………3 { we assign an artificial coordinate x0 =1 to obtain limits from 0 to d}

#### in vector form  :

h(x) = sign(wTx) .. 4 { designates the hypothesis set}

So , in short what does all this mean ??

Given a training set , pick up a misclassified point and adjust the weights to get a proper classification . so if sign(wTx) !=y then adjust weight w as w+yx ..

#### After all this mathematics.. moving a bit closer to the types of learning currently available :

- Supervised learning : Given( input, correct output) predict for future inputs e,g a coin vending machine learns how much each coin weighs, after looking at 1000000 inputs.. say u put 1s ,2s, 5s and 10s coins in the vending machine.After that any coin u dump in the vending machine, it will classify the coin as 1s, 2s , 5s or 10s.

---

- Unsupervised Learning : Given( input) predict for future inputs  ( How is this possible??).. Well.. You dont know the coin denomination yet but u know that they are of different weights..so u just classify it as type1,type2, type4 and type4 .. someday ..someone will tell u they are 1s,2s, 5s and 10s .. and all that  you have to do is assign the type to the denomination.

---
